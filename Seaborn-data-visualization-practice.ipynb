## Visualization - Scatter Plots

The scatter plot is a mainstay of statistical visualization. It depicts the joint distribution of two variables using a cloud of points, where each point repersents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

Although, scatter plot can be plotted for both numerical as well as categorical data, its mostly used to depict relationshipbetween numarical data.

There are two ways to draw scatterplot in seaborn
- 1. relplot(Figure Level Fuction)
- 2. scatterplot(Axes Level Function)

As we have seen these visualization can be quite helpful because they use simple and easily understood repersentations of data even on complex data by mapping up to three additional variables using the semantics of hue, size and style.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.style.use("fivethirtyeight")
```

```python
data=sns.load_dataset('tips')
data.head()
```

## Plot a scatterplot between total_bill and tip

```python
sns.relplot(x='total_bill', y='tip', data=data)
```

```python
sns.relplot(x='total_bill', y='tip',kind='scatter', data=data)
```

## Add hue parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

## Add style parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex', data=data)
```

## Add the size parameter

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex', size='size', data=data)
```

```python
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex', size='size',sizes=(15,200), data=data)
```

## Using the scatter fucntion

```python
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=data)
```

```python
sns.catplot(x='day', y='tip', kind= 'strip', data=tips)  # use jitter
```

```python
sns.catplot(x='day', y='tip', jitter=0, kind= 'strip', data=tips)
```

```python
sns.catplot(x='day', y='tip', kind= 'swarm', data=tips)  # use jitter
```

```python
sns.catplot(x='day', y='tip', kind= 'swarm', hue='sex', data=tips)
```

```python
sns.swarmplot(x='day', y='tip', hue='sex', data=tips)
```

```python
sns.swarmplot(x='day', y='tip', data=tips)
```

## Violinplots

Violin plots are a method of plotting numerical data and can be considered  a combination of the box plot with a kernel density plot.

```python
sns.violinplot(tips["total_bill"])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', hue='sex', kind='violin', split=True, data=tips)
```

## Visualization - Heatmaps

A heatmap is a graphical repersentation of data where the individual values contained in a matrix are repersented as colors

```python
flights=sns.load_dataset('flights')
flights.head()
```

```python
x = flights.pivot_table(index='year', columns='month', values='passengers', aggfunc='sum')
x
```

```python
sns.heatmap(x)
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```

## Clustermap

A heatmap is a graphical repersentation of data where the individual values contained in a matrix are repersented as colors

```python
sns.clustermap(x)
```

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, metric='correlation', annot=True)
```

```python
sns.clustermap(x, z_score=0, annot=True, row_cluster=False, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, col_cluster=False, metric='correlation')
```

## Visualization - Joint Plot

Jointplots creates a multi-panel figure that shows both the bivariate(or joint) relationship between two variables along with the univariate (or marginal) distribution of each on seperate axes.

We acn use jointplots in the following manners:
1. Scatter Plots
2. Hexbin Plots
3. Kde Plots
4. Reg Plots

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips )
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips )
```

```python
sns.jointplot(x='total_bill', y='tip', kind='kde', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

## Visualization - Pair Plot

Pairplots are used to plot multiple pairwise bivariate distribution in a dataset. This creates a matrix of axes and shows the relationship for each pair of columns in a dataframe. By default, it is also draws the univariate distribution of each variable on the diagonal axes.

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris, hue='species')
```

```python
plt.figure(figsize=(8,10))
plt.subplot(121)
sns.swarmplot('species', 'petal_length', data=iris);
plt.subplot(122)
sns.violinplot('species', 'petal_length', data=iris);
```

## Visualization - DistPlot

DistPlots are used to do univariate analysis on a continous/numerical data.

Distplot combines 3 things in one graph:

1. Histogram
2. Kernael Density Estimation(PDF)
3. Rugplot

```python
titanic=sns.load_dataset('titanic')
titanic.head()
```

```python
titanic['age'].fillna(titanic['age'].mean(), inplace=True)
```

## Distplot

```python
sns.distplot(titanic['age'])
```

## Only Histogram

```python
sns.distplot(titanic['age'], kde=False)
```

```python
sns.histplot(titanic['age'])
```

```python
sns.distplot(titanic['age'], bins=10, kde=False)
```

```python
sns.distplot(titanic['age'], bins=100, kde=False)
```

## Only kde

```python
sns.distplot(titanic['age'], hist=False)
```

```python
sns.kdeplot(titanic['age'])
```

## Rugplot

```python
sns.distplot(titanic['age'], hist=False, kde=False, rug=True)
```

# Data Visualization - using Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
%matplotlib inline
```

```python
# see the pre-defined styles provided.
plt.style.available
```

```python
# use the 'seaborn-colorblind' style
plt.style.use('seaborn-colorblind')
```

```python
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0), 
                   'B': np.random.randn(365).cumsum(0) + 20,
                   'C': np.random.randn(365).cumsum(0) - 20}, 
                  index=pd.date_range('1/1/2017', periods=365))
df.head()
```

```python
df.plot(); # add a semi-colon to the end of the plotting call to suppress unwanted output
```

```python
df.plot('A','B', kind = 'scatter');
```

You can also choose the plot kind by using the DataFrame.plot.kind methods instead of providing the kind keyword argument.

kind :

- 'line' : line plot (default)
- 'bar' : vertical bar plot
- 'barh' : horizontal bar plot
- 'hist' : histogram
- 'box' : boxplot
- 'kde' : Kernel Density Estimation plot
- 'density' : same as 'kde'
- 'area' : area plot
- 'pie' : pie plot
- 'scatter' : scatter plot
- 'hexbin' : hexbin plot

```python
# create a scatter plot of columns 'A' and 'C', with changing color (c) and size (s) based on column 'B'
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
```

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal')
```

```python
df.plot.box();
```

```python
df.plot.hist(alpha=0.7);
```

```python
df.plot.kde();
```
